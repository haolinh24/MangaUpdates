## Genres distributions between publisher and scanlator

```python
def genres_publisher_vs_scanlator(name_of_publisher):
    df_publisher = df_big_publisher[df_big_publisher['name of English publisher'].str.contains(name_of_publisher)]
    df_publisher_genres = df_publisher[['Action','Adult','Adventure','Comedy','Doujinshi','Drama','Ecchi','Fantasy','Gender Bender','Harem','Hentai','Historical','Horror','Josei','Lolicon','Martial Arts','Mature','Mecha','Mystery','Psychological','Romance','School Life','Sci-fi','Seinen','Shotacon','Shoujo','Shoujo Ai','Shounen', 'Shounen Ai','Slice of Life','Smut','Sports','Supernatural','Tragedy','Yaoi','Yuri']]
    df_publisher_genres_count = df_publisher_genres.sum(axis=0).to_frame()
    df_publisher_genres_count.columns = ['Number of titles by publisher']
    df_scanlator = df_publisher[(df_publisher['Number of Scanlation group']!=0) & (df_publisher['Digitally published first by publisher']==False)]
    df_scanlator_genres = df_scanlator[['Action','Adult','Adventure','Comedy','Doujinshi','Drama','Ecchi','Fantasy','Gender Bender','Harem','Hentai','Historical','Horror','Josei','Lolicon','Martial Arts','Mature','Mecha','Mystery','Psychological','Romance','School Life','Sci-fi','Seinen','Shotacon','Shoujo','Shoujo Ai','Shounen', 'Shounen Ai','Slice of Life','Smut','Sports','Supernatural','Tragedy','Yaoi','Yuri']]
    df_scanlator_genres_count = df_scanlator_genres.sum(axis=0).to_frame()
    df_scanlator_genres_count.columns = ['Number of scanlated titles']
    df_scanlator_vs_publisher = pd.concat([df_publisher_genres_count, df_scanlator_genres_count], axis=1)
    df_scanlator_vs_publisher['Scanlation percentage'] = (df_scanlator_vs_publisher['Number of scanlated titles']/df_scanlator_vs_publisher['Number of titles by publisher']*100).round(2)
    return df_scanlator_vs_publisher

     

    
    
```

```python
df_publisher_genres = df_big_publisher[['Action','Adult','Adventure','Comedy','Doujinshi','Drama','Ecchi','Fantasy','Gender Bender','Harem','Hentai','Historical','Horror','Josei','Lolicon','Martial Arts','Mature','Mecha','Mystery','Psychological','Romance','School Life','Sci-fi','Seinen','Shotacon','Shoujo','Shoujo Ai','Shounen', 'Shounen Ai','Slice of Life','Smut','Sports','Supernatural','Tragedy','Yaoi','Yuri']]
df_scanlator_genres = df_big_publisher_scanlate[['Action','Adult','Adventure','Comedy','Doujinshi','Drama','Ecchi','Fantasy','Gender Bender','Harem','Hentai','Historical','Horror','Josei','Lolicon','Martial Arts','Mature','Mecha','Mystery','Psychological','Romance','School Life','Sci-fi','Seinen','Shotacon','Shoujo','Shoujo Ai','Shounen', 'Shounen Ai','Slice of Life','Smut','Sports','Supernatural','Tragedy','Yaoi','Yuri']]
df_publisher_genres_count = df_publisher_genres.sum(axis=0).to_frame()
df_scanlator_genres_count = df_scanlator_genres.sum(axis=0).to_frame()
df_scanlator_vs_publisher = pd.concat([df_publisher_genres_count, df_scanlator_genres_count], axis=1)
df_scanlator_vs_publisher.columns = ['Number of titles', 'Number of scanlated titles']
df_scanlator_vs_publisher['Scanlation percentage'] = (df_scanlator_vs_publisher['Number of scanlated titles']/df_scanlator_vs_publisher['Number of titles']*100).round(2)
# df_scanlator_vs_publisher.sort_values(by=['Scanlation percentage'], ascending=False)
df_scanlator_vs_publisher['Genres percentage'] = (df_scanlator_vs_publisher['Number of titles']/1534*100).round(2)
df_scanlator_vs_publisher.sort_values(by=['Genres percentage'], ascending=False)
```

```python
def who_got_most_scanlated(name_of_publisher):
    df_publisher = df_big_publisher[df_big_publisher['name of English publisher'].str.contains(name_of_publisher)]
    df_scanlator = df_publisher[(df_publisher['Number of Scanlation group']!=0) & (df_publisher['Digitally published first by publisher']==False)]
    list_publisher_vs_scanlator = [len(df_publisher),len(df_scanlator)]
    return list_publisher_vs_scanlator
```

```python
scanlator_vs_publisher = {'Viz':who_got_most_scanlated('Viz'), 'Yen Press':who_got_most_scanlated('Yen Press'), 'Seven Seas':who_got_most_scanlated('Seven Seas'), 'Kodansha':who_got_most_scanlated('Kodansha')}
```

```python
df_scanlate_by_publisher = pd.DataFrame.from_dict(scanlator_vs_publisher).T
df_scanlate_by_publisher.columns = ['Number of titles', 'Number of scanlated titles']
df_scanlate_by_publisher['Scanlation Percentage'] = (df_scanlate_by_publisher['Number of scanlated titles']/df_scanlate_by_publisher['Number of titles']*100).round(2)
df_scanlate_by_publisher
```
