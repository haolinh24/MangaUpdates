```python
df_scanlator_date_new = df_scanlator_date[['Name','First Update New','Latest Update New']]
```

```python
df_scanlator_date_new[(df_scanlator_date_new['First Update New']<datetime.date(2005,12,31))&(df_scanlator_date_new['Latest Update New']>datetime.date(2005,1,1))]
```

```python
dict_scanlator_year = {}
for i in range(2005, 2021):
    dict_scanlator_year[i] = 0
```

```python
dict_scanlator_year
```

```python
for j in range(2005, 2021):
    df_scanlator_year_count = df_scanlator_date_new[(df_scanlator_date_new['First Update New']<datetime.date(j,12,31))&(df_scanlator_date_new['Latest Update New']>datetime.date(j,1,1))]
    dict_scanlator_year[j] = len(df_scanlator_year_count)
```

```python
dict_scanlator_year
```

```python
df_scanlator_year = pd.DataFrame.from_dict(dict_scanlator_year, orient='index')
```

```python
All_Genres = ['Action','Adult','Adventure','Comedy','Doujinshi','Drama','Ecchi','Fantasy','Gender Bender','Harem','Hentai','Historical','Horror','Josei','Lolicon','Martial Arts','Mature','Mecha','Mystery','Psychological','Romance','School Life','Sci-fi','Seinen','Shotacon','Shoujo','Shoujo Ai','Shounen', 'Shounen Ai','Slice of Life','Smut','Sports','Supernatural','Tragedy','Yaoi','Yuri']
dict_genres_distribution = {}
for i in All_Genres:
    dict_genres_distribution[i] = 0
```

```python
for i in All_Genres:
    df_genres = df_scanlator[df_scanlator[i] != 0]
    dict_genres_distribution[i] = len(df_genres)
```

```python
dict_genres_distribution
```

```python
df_genres_distribution = pd.DataFrame.from_dict(dict_genres_distribution, orient='index')
```
