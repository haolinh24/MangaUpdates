# Kodansha requests

```python
df_kodansha_request = df_all_request[df_all_request['Copyright owner name'].str.contains('Kodansha')|df_all_request['Copyright owner name'].str.contains('KODANSHA')]
```

```python
df_kodansha_request
```

```python
df_kodansha_request['URLs removed'].sum()
```

```python
df_kodansha_request.sort_values(by='Date')
```

```python
df_kodansha_request['New Date'] = pd.to_datetime(df_kodansha_request['Date']).dt.date
```

```python
kodansha_request_dict = {}
```

```python
for x in range(2013, 2021):
    kodansha_request_year = df_kodansha_request[(df_kodansha_request['New Date'] >= datetime.date(x, 1, 1)) & (df_kodansha_request['New Date'] <= datetime.date(x, 12, 31))]
    kodansha_request_number = kodansha_request_year['URLs removed'].sum()
    kodansha_request_dict[x] = kodansha_request_number                                            
```

```python
kodansha_request_dict
```
