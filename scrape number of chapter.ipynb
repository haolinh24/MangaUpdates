```python
# Get the name of the group
def get_name(soup):
    GroupName = soup.find_all("div", "p-1 col-6 specialtext")
    return GroupName[0].text
```

```python
# Group has the unknown start day or not
def unknown_start_day(link_group, soup):
    link_last_page = link_group+'&page='+number_page(soup)
    req_last_page = myGet(link_last_page)
    soup_last_page = bs(req_last_page.content)
    date_list = soup_last_page.find_all('div','col-2 text')
    if date_list == []:
        return "No release"
    else:
        if date_list[len(date_list) - 1].text == "Unknown":
            return True
        else:
            return False
```

```python
# Get the number group's release pages
def number_page(soup):
    if soup.find_all("span", "d-none") == []:
        page_series_number = "1"
    else:
        total_page_unclean = soup.find_all("span","d-none")[0].text.strip()
        page_series_number = re.findall(r"[0-9]+",total_page_unclean)[0]
    return page_series_number
        
```

```python
def chapter_year_element(link_group, soup, dict_group_release):
    for i in range(1, int(number_page(soup))+1):
        link_page = link_group+'&page='+str(i)
        req_link_page = myGet(link_page)
        soup_link_page = bs(req_link_page.content)
        release_date = soup_link_page.find_all("div","col-2 text") 
        group_and_title = soup_link_page.find_all('div','col-4 text')
        group_and_title_text = [i.text for i in group_and_title]
        #some chapters were done by more than 1 group  
        #we only count it as numbers of chapter/numer of groups
        group_text = []
        for i in range(len(group_and_title_text)): 
            if i%2==1:
                group_text.append(group_and_title_text[i])
        group_seperate = [i.split(' & ') for i in group_text]
        for x in range(len(release_date)):
            if release_date[x].text != 'Unknown':
                for year in list_year:
                    if year == int('20'+release_date[x].text[-2:]):
                        dict_group_release[year] +=(1/len(group_seperate[x]))
    return dict_group_release
```

```python
def chapter_year(link_group, soup):
    for year in list_year:
        dict_group_release[year] = 0 #start with the new, blank dictionary
    if unknown_start_day(link_group, soup) == 'No release':
        return dict_group_release
    if unknown_start_day(link_group, soup) == True:
        Unknown = 0
        year_start = ''
        for i in range(int(number_page(soup)), 1, -1):
            link_page = link_group+'&page='+str(i)
            req_link_page = myGet(link_page)
            soup_link_page = bs(req_link_page.content)
            date_list = soup_link_page.find_all('div','col-2 text')
            if date_list[len(date_list) - 1].text == "Unknown":
                for i in range(len(date_list)-1,0,-1):
                    if date_list[i].text == "Unknown":
                        Unknown +=1
                    else:
                        year_start = "20"+date_list[i].text[-2:]
                        break
        if year_start == '':
            return dict_group_release
        else:
            dict_group_release[int(year_start)] += Unknown
            return chapter_year_element(link_group, soup, dict_group_release) 
       
    else:
        return chapter_year_element(link_group, soup, dict_group_release)
            
    
    
```

```python
def group_cell(link_group):
    req_group = myGet(link_group)
    soup_group = bs(req_group.content)
    group_dict = {"Name":get_name(soup_group),"Unknown start day":unknown_start_day(link_group, soup_group)}
    group_dict.update(chapter_year(link_group,soup_group))
    group_cell = pd.DataFrame.from_dict(group_dict, orient='index').transpose()
    return group_cell
```

```python
# function to create a dataframe for a 25-group-page
```

```python
def scan_each_page(page_link):
    df_group_page = pd.DataFrame()
    req_group_page = myGet(page_link)
    soup_group_page = bs(req_group_page.content.decode('utf-8', 'ignore'))
    scanlator_link_include = soup_group_page.find_all("a", alt="Group Info")
    scanlator_link = [i.get('href') for i in scanlator_link_include]
    for link in scanlator_link:
        df_group_page = df_group_page.append(group_cell(link))
    return df_group_page
```
